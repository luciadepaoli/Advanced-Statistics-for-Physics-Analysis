Produce a plot with the Oil price as a function of the year.

```r
options(repr.plot.width=12, repr.plot.height=8)

ggplot(crude_oil, aes(x=Year, y=OilPrice)) + geom_point(color="firebrick", size=2) +labs(y='Oil price') +
        geom_line(color='firebrick') + ggtitle("Oil price during the years")+
        theme(plot.title = element_text(hjust = 0.5, size=15), axis.text=element_text(size=9), 
          axis.title.x = element_text(size = 12), 
          axis.title.y = element_text(size = 12))
```

### 2.4
Plot the derivative of the curve, simply evaluated with the finite difference formula (forward derivative):

```r
Deriv <- NULL
for(i in 1:length(crude_oil$OilPrice)){
    
    Deriv <- c(Deriv, crude_oil$OilPrice[i+1] - crude_oil$OilPrice[i])
    
    }
Deriv <- as.numeric(unlist(Deriv))

crude_oil <- cbind(crude_oil, Deriv)

ggplot(crude_oil, aes(x=Year, y=Deriv))+ geom_point(color="firebrick", size=3) +labs(y='Derivatives of oil price') +
    geom_line(color='firebrick') + ggtitle("Derivatives of oil price") +
    theme(plot.title = element_text(hjust = 0.5, size=20), axis.text=element_text(size=10), 
          axis.title.x = element_text(size = 15), 
          axis.title.y = element_text(size = 15))
```

### 3.1
Write R code that is able to read the file and import it in a tibble structure.

I have removed all the rows with empty Code because there are aggregation of countries.

```r
id <- '10kgRQGKEC81QoVW9kj8goNzzyLcT0xld'
coal <- read.csv(sprintf("https://docs.google.com/uc?id=%s&export=download", id),skip=0)
library(tibble)
coal_tibb_orig <- as_tibble(coal)
coal_tibb <- filter(coal_tibb_orig, Code!='' & Entity!='World')
```

### 3.2
Count the number of countries available in the file and produce a barplot with the number of entries for each country.

```r
uniq <- distinct(coal_tibb, Entity)
cat('Number of countries available: ',nrow(uniq))

options(repr.plot.width=10, repr.plot.height=30)

ggplot(coal_tibb, aes(Entity)) + coord_flip() +
labs(y='# entries', x='Country') + ggtitle("# Entries for each country") +
theme(plot.title = element_text(hjust = 0.5, size=10), axis.text=element_text(size=10), 
        axis.title.x = element_text(size = 10), 
        axis.title.y = element_text(size = 10)) +
geom_bar(fill="firebrick") 
```

for the following items select only the years $>= 1970$:

```r
coal_cut <- filter(coal_tibb, Year >= 1970)
uniq_cut <- distinct(coal_cut, Entity,.keep_all = T)
```

### 3.3
Selecting only the year after 1970, determine the total integrated production for each country and print the top 5 Countries with highest coal productions.

```r
selected <- group_by(coal_cut, Entity) %>% summarise(Total = sum(Coal.production..TWh.)) %>%
arrange(desc(Total)) %>% filter(row_number()==1:5)
selected
```

### 3.4
For the 5 top Countries, create a plot of production as a function of time.

```r
df <- filter(coal_tibb, Entity == 'China' | Entity == 'United States' | Entity == 'Russia' | Entity == 'Australia' | 
      Entity == 'India')
options(repr.plot.width=12, repr.plot.height=8)

ggplot(df, aes(x=Year, y=Coal.production..TWh., group=Entity, col=Entity)) + geom_point(size=2) +
labs(y='Coil production') +
    geom_line() + ggtitle("Coil production top 5 countries") +
    theme(plot.title = element_text(hjust = 0.5, size=15), axis.text=element_text(size=9), 
          axis.title.x = element_text(size = 12), 
          axis.title.y = element_text(size = 12))
```

### 3.5
Generate a plot with the cumulative sum of the World's coal production over the years.

```r
cum <- filter(coal_tibb_orig, Entity == 'World') %>% 
group_by(Entity, Year) %>% summarise(Total = cumsum(Coal.production..TWh.))

ggplot(cum, aes(x=Year, y=Total, group=Entity, col=Entity)) + geom_point(size=2,color='firebrick') +
labs(y='Cumulative sum') +
    geom_line(color='firebrick') + ggtitle("Cumulative sum of World's coal production over the years") +
    theme(plot.title = element_text(hjust = 0.5, size=15), axis.text=element_text(size=9), 
          axis.title.x = element_text(size = 12), 
          axis.title.y = element_text(size = 12))
```
